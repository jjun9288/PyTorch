## 1. One-Hot Encoding

One-hot encoding은 선택해야 하는 선택지의 갯수만큼 차원을 가지면서 각 선택지의 인덱스에 해당하는 원소에는 1, 나머지 원소에는 0의 값을 가지도록 표현하는 방법이다.   
예를 들어 강아지, 고양이, 냉장고 3개의 선택지가 있다면, 

```python
강아지 = [1,0,0]
고양이 = [0,1,0]
냉장고 = [0,0,1]
```

정수 인코딩 (1,2,3,4,...)이 아닌 원핫 인코딩(0,1,0,0,...) 을 쓰는 이유는 오차의 균등함을 위해서이다. 각각 MSE를 
구했을 때 원-핫 벡터는 모든 쌍에 대해 유클리드 거리를 구했을 때 유클리드 거리가 전부 동일하게 나온다.   
이를 우리는 원-핫 벡터의 클래스 표현 방법이 '무작위성을 가진다'고 한다.   
하지만 이 무작위성은 유사성을 구할 수 없다는 단점으로 언급되기도 한다.

## 2. Softmax Regression

이전 Logistic Regression에서는 이진 분류(Binary Classification)을 풀었다면,   
Softmax Regression에서는 다중 클래스 분류(Multi-Class Classification)을 푼다.

#### 1. 다중 클래스 분류

Logistic Regression에서 Sigmoid function은 예측값을 0과 1 사이로 만든다. 예를 들어 스팸 메일 분류기를 Logistic Regression으로 구현했을 때 출력이 0.75라면 이 이메일이 스팸일 확률이 0.75라는 것이다. 반대로 스팸이 아닐 확률은 25%가 되는 것이다.



Softmax Regression은 확률의 총 합이 1이 되는 이 아이디어를 다중 클래스 분류 문제에 적용한다. 



#### 2. Softmax function

Softmax function은 분류해야하는 클래스의 총 갯수를 k라 할 때, k차원의 벡터를 입력받아 각 클래스에 대한 확률을 추정한다.

##### Softmax 함수의 입력으로 어떻게 바꿀까?

하나의 샘플 데이터에는 4개의 독립 변수(SepalLength, SepalWidth, PetalLength, PetalWidth)를 갖는다.   
하지만 Softmax 함수의 입력으로 사용되는 벡터는 분류하고자 하는 클래스(virginica, setosa, versicolor)의 갯수가 되어야 하므로 3차원 벡터로 변환되어야 한다.



이를 위해서는 샘플 데이터를 소프트맥스 함수의 입력 벡터로 차원을 축소해야 한다. 위의 그림 같은 경우에슨 (4x3) matrix를 이용하면 될 것이다.

##### 오차를 어떻게 구할까?

Softmax 회귀에서는 실제값을 One-hot vector로 표현한다. 이와 예측값의 오차를 구한 후 이 오차로부터 가중치를 업데이트를 해나가면 된다.
